## Creating train_test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.33,random_state=5)
```

```python
X_train.shape
```

## Feeding our Dataset to our LinearRegression Model

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
def adj_r2(x,y):
    r2=lm.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
#Training Accuracy Score Prediction
y_pred=lm.predict(X_train)
adj_r2(X_train,y_train)
```

```python
print('Training Accuracy: ',lm.score(X_train,y_train)*100)
```

```python
#Predicting the test Accuracy
print("Test accuracy : ",lm.score(X_val,y_val)*100)
```

```python
lm.predict(X_val)
```

```python
#Adjusted R2 for test dataset
adj_r2(X_val,y_val)
```

### Ridge Regularization

```python
#Ridge regression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
ridge = linear_model.Ridge()
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_val)
print(mean_squared_error(y_pred_ridge, y_val))


print(ridge.coef_)
print(ridge.intercept_)
```

```python
from sklearn.linear_model import RidgeCV
alphas=np.random.uniform(low=0,high=10,size=(50,))
ridgecv=RidgeCV(alphas=alphas,cv=10,normalize=True)
ridgecv.fit(X_train,y_train)
```

```python
## Let us create a Ridge based Regression Model since we got our optimal parameter alpha as 0.428375
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(X_train,y_train)
```

### Lasso Regularization

```python
#Lasso Regularization
#Here we are doing Lasso for 10 cross validation datas
from sklearn.linear_model import LassoCV
lasscv=LassoCV(alphas=None,cv=100,max_iter=100000,normalize=True)
lasscv.fit(X_train,y_train)
```

```python
lasscv.alphas_
```

```python
#Best alpha value returned by LassoCV
alpha=lasscv.alpha_
alpha
#Got alpha = 0.0003592777
```

```python
#Now that we have best parameter, let us go for Lasso Regression
lasso_reg=Lasso(alpha)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_train,y_train)*100
```
